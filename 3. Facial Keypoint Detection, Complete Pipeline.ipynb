```python
# From Udacity's Computer Vision Course
def crop_to_roi(image, x, y, h, w, n_out):
    image_copy = np.copy(image)
    if h < n_out or w < n_out:
        d = max(n_out - h, n_out - w)
        pad = int(d//2)
    else:
        pad = 0
    roi = image_copy[y-pad:y+h+pad, x-pad:x+w+pad]
    roi = cv2.resize(roi, (n_out, n_out))
    return roi

def normalize(image):
    roi = np.copy(image)
    roi = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

    ## TODO: Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
    roi = roi / 255.0
    return roi

def to_tensor(image):
    roi = np.copy(image)
    roi = roi.reshape(roi.shape[0], roi.shape[1], 1)
    roi = roi.transpose((2,0,1))
    roi_tensor = torch.from_numpy(roi).unsqueeze(dim=0).type(torch.FloatTensor)
    return roi_tensor
```

```python
# From earlier - Udacity
def show_all_keypoints(image, predicted_key_pts):
    """Show image with predicted keypoints"""
    # image is grayscale
    plt.imshow(image, cmap='gray')
    plt.scatter(predicted_key_pts[:, 0], predicted_key_pts[:, 1], s=20, marker='.', c='m')

# From earlier - Udacity
# visualize the output
# by default this shows a batch of 10 images
def visualize_output(test_image, test_output):

    plt.figure(figsize=(20,10))
    ax = plt.subplot(1, 1, 1)

    # un-transform the image data
    image = test_image.data   # get the image from it's Variable wrapper
    img = image.numpy()   # convert to numpy array from a Tensor
    # image = np.transpose(img, (1, 2, 0))   # transpose to go from torch to numpy image

    # un-transform the predicted key_pts data
    predicted_key_pts = test_output.data
    predicted_key_pts = predicted_key_pts.numpy().squeeze()
    # undo normalization of keypoints  
    predicted_key_pts = predicted_key_pts*50.0+100
    
    # call show_all_keypoints
    show_all_keypoints(np.squeeze(img), predicted_key_pts)
        
    # plt.axis('off')

    plt.show()
```
